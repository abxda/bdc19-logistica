```python
df=pd.read_csv("../data/logistica-gas.csv", parse_dates=['fecha'], encoding='UTF-8')
```

```python
df['km_total'] = df['km']-df['kmAnterior']
df['km_x_litro'] = df['km_total']/df['volumen']
```

```python
df['rubro']=df['rubro'].replace(['BP REGULAR', 'AKRON REGULAR', 'REPSOL EFITEC 87', 'SHELL SUPER','G SUPER','MOBIL EXTRA','PREMIUM'], 'MAGNA')
df['rubro']=df['rubro'].replace(['BP DIESEL', 'SHELL DIESEL', 'DIESEL2', 'DIESEL 2'], 'DIESEL')
df['rubro']=df['rubro'].replace(['GAS NATURAL'], 'GAS LP')
```

```python
df_total_volumen = df.groupby(['subagencia', 'rubro', 'fecha'])[['total','volumen']].sum().reset_index()
```

```python
df_total_volumen = df.groupby(['subagencia', 'rubro', 'fecha'])[['total','volumen','precioLitro']].agg([('csum', 'sum'), ('cmin', 'min')]).reset_index()
```

```python
#df_total_volumen.to_csv("/home/abxda/entrega/bdc19-logistica/data/df_total_volumen.csv")
```

```python
df_total_volumen['mxn_x_litro'] = df_total_volumen[('total', 'csum')]/df_total_volumen[('volumen', 'csum')]
```

```python
df_total_volumen['total_min'] = df_total_volumen[('precioLitro', 'cmin')]*df_total_volumen[('volumen', 'csum')]
```

```python
df_total_volumen['ahorro'] = df_total_volumen[('total', 'csum')]-df_total_volumen['total_min']
```
